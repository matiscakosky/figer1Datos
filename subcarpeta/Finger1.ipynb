```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('events.csv' ,low_memory = False)
```

# Punto 1: Tipo de evento predominante en el set de Datos

```python
eventoPred = df['event'].value_counts().head(5)
plt1 = eventoPred.plot(kind = 'bar', grid=True, title = "Eventos de Usuarios", figsize = (20,15), legend=True, fontsize = 12)
plt1.set_xlabel("Eventos", fontsize = 18)
plt1.set_ylabel("Cantidad de eventos", fontsize = 18)
```

### El evento predominante en el set de datos es "Viewed product"

# Punto 2: Encontrar los 5 productos mas vendidos

```python
dfVentas = df.loc[df['event']=='conversion',['event','sku','model','condition','storage','color']]
dfVentas['descrpitionCompleted'] = dfVentas['sku'] + ' - ' + dfVentas['model'] + ' ' + dfVentas['storage'] + ' ' + dfVentas['color']
dfVentasPorModelo = dfVentas['descrpitionCompleted'].value_counts().head(5)

plt2 = dfVentasPorModelo.plot(kind = 'barh', grid=True, title = "Productos mas exitosos", figsize = (20,15), legend=True, fontsize = 20)
plt2.set_xlabel("Cantidad de Ventas", fontsize = 18)
plt2.set_ylabel("Modelos", fontsize = 18)

```

### El modelo mas vendido es el Samsung Galaxy J5 16Gb en Dorado

# Punto 3: Hallar los 10 usuarios que registraron mas actividad en la plataforma
